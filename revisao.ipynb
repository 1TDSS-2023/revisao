### ### **Arrays**

* Array de strings

```js

```

* Array de arrays

```js
let alunos = [
    ["Lucas", 23, 80.0, true],
    ["João", 25, 70.0, true],
    ["Maria", 20, 60.0, false]
];

// Acessando o primeiro aluno
alunos[0]; // ["Lucas", 23, 80.0, true]

// Acessando o nome do primeiro aluno
alunos[0][0]; // "Lucas"
```

## ### **JSON**

* Declaração

```js
let aluno = {
    "nome": "Lucas",
    "idade": 23,
    "peso": 80.0,
    "humano": true
};

// Acessando o nome do aluno
aluno.nome; // "Lucas"
aluno["nome"]; // "Lucas"

// Array de JSON

let alunos = [
    {
        "nome": "Lucas",
        "idade": 23,
        "peso": 80.0,
        "humano": true
    },
    {
        "nome": "João",
        "idade": 25,
        "peso": 70.0,
        "humano": true
    },
    {
        "nome": "Maria",
        "idade": 20,
        "peso": 60.0,
        "humano": false
    }
];

// Acessando o primeiro aluno
alunos[0]; // {"nome": "Lucas", "idade": 23, "peso": 80.0, "humano": true}

// Acessando o nome do primeiro aluno
alunos[0].nome; // "Lucas"

// Acessando o nome do primeiro aluno
alunos[0]["nome"]; // "Lucas"
```

### ### **Acessando JSON de maneira recursiva**

```js
//Imprimir array de objetos e suas propiedades utilizando forEach e template string.
alunos.forEach(aluno => {
    console.log(`${aluno.nome} - ${aluno.nota} - ${aluno.disciplina}`);
});
```
